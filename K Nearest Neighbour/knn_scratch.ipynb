```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from collections import Counter
from sklearn.metrics import accuracy_score

data = load_iris()
X=data.data
Y=data.target

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

def calEuclideanDistance(p1,p2):
    return np.sqrt(np.sum((np.array(p1)-np.array(p2))**2))

def kNearestNeighbour(X_train,Y_train,X_test,k):
    predictions=[]
    for test_point in X_test:
        distances=[]
        for i in range(len(X_train)):
            dis = calEuclideanDistance(test_point,X_train[i])
            distances.append((dis,Y_train[i]))
        distances.sort(key = lambda x: x[0])
        k_nearest =distances[:k]
        labels = [label for _,label in k_nearest]
        mostFreq = Counter(labels).most_common(1)[0][0]
        predictions.append(mostFreq)
    return predictions


for k in range(1,10,2):
    Y_pred = kNearestNeighbour(X_train=X_train,Y_train=Y_train,X_test=X_test,k=k)
    acc = accuracy_score(Y_test,Y_pred)
    print(f"K={k} -> Accuracy={acc}")
```
