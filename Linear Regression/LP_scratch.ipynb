```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('student_Marks.csv')
print(df)
print(df.info())

print(df.corr())

new_df=df.drop(columns=["number_courses"])
print(new_df.to_string())

x=df['time_study'].values
y=df['Marks'].values

x_mean=np.mean(x)
y_mean=np.mean(y)

num=np.sum((x-x_mean)*(y-y_mean))
den=np.sum((x-x_mean)*(x-x_mean))
m=num/den

c=y_mean-m*x_mean

print("Slope is: ",m)
print("Intercept is: ",c)

def predict_output(time):
    return m*time+c

mse=np.mean((y-predict_output(x))**2)

print("Mean Sqaure Error is: ",mse)

plt.plot(x,predict_output(x))
plt.scatter(x,y)
plt.xlabel("time")
plt.ylabel("Marks")
plt.title("Linear Regression")
plt.show()
```
